```python
import single_min_tp as smt
import networkx as nx
import random as rn
import glob
import create_grid as cg
```

```python
n = 5
graph_files = glob.glob('./graphs/grid{}/*.graphml'.format(n))
file_name = 'grid_{}_{}.csv'.format(n, n)
repeat = 5
priority_nodes = [4, 5, 6, 7]
count = 0
for _ in range(repeat):
    for p in priority_nodes:
        pn = rn.sample(list(map(str, range(n**2))), p)
        for graph_path in graph_files:
            graph_name = list(graph_path.split('/'))[-1].split('.')[0]
            g = nx.read_graphml(graph_path)
            c = smt.TSP_CONCAT(g, graph_name, pn)
            c.solve()
            c.output(file_name, str(count))
            count += 1
            print(count)
```
